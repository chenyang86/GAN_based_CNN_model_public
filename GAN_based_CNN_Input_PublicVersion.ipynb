```python
# Functions
def Loc(x,y,a,b):
    # Enter two pairs of longitude and latitude to restore the horizontal and vertical distance
    x1 = (a-x)/180*np.pi*meter_R
    y1 = (b-y)/180*np.pi*meter_R 
    
    return x1,y1
```
